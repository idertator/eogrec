```python
X, H, V = [], [], []
with open('./test1.txt') as f:
    for l in f:
        if l != '\n':
            x, h, v = l.strip().split()
            X.append(int(x))
            H.append(int(h))
            V.append(int(v))

plot(X, H)
```
